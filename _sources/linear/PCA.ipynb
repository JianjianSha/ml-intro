# 主成分分析

主成分分析基于方差最大准则，使得数据在这些方向上的映射最大程度的体现原分布情况。记数据集 $D=\{\mathbf t_n \in R^D: n=1,2, \cdots , N\}$，记最具有代表性的主成分方向为 $\mathbf v_1$，令 $\mathbf v_1$ 为单位向量，即 $\mathbf v_1^{\top} \mathbf v_1=1$，某数据点 $\mathbf t$ 在 $\mathbf v_1$ 方向上的投影为 $\mathbf v_1^{\top} \mathbf {t \cdot v_1}$，且 $\mathbf v_1^{\top} \mathbf {t}$ 是标量，对其转置其值不变，那么这个数据点与中心 $\overline {\mathbf t}$ 在 $\mathbf v_1$ 方向上投影的平方差为

$$\|\mathbf v_1^{\top}  \mathbf {t \cdot v_1} - \mathbf v_1^{\top} \mathbf {\overline t \cdot v_1}\|^2=(\mathbf v_1^{\top}  \mathbf t-\mathbf v_1^{\top}  \mathbf {\overline t})^2 \|\mathbf v\|^2=(\mathbf v_1^{\top}  \mathbf t-\mathbf v_1^{\top}  \mathbf {\overline t})(\mathbf t^{\top}  \mathbf v_1- \mathbf {\overline t}^{\top} \mathbf v_1)=\mathbf v_1^{\top}(\mathbf t- \mathbf {\overline t})(\mathbf t - \mathbf {\overline t})^{\top} \mathbf v_1$$

所以数据集在映射空间的方差为

$$\begin{aligned} V_1&=\frac 1 N \sum_{n=1}^N(\mathbf v_1^{\top}  \mathbf t_n-\mathbf v_1^{\top}  \mathbf {\overline t})^2 
\\ &=\mathbf v_1^{\top}\left( \frac 1 N \sum_{n=1}^N (\mathbf t- \mathbf {\overline t})(\mathbf t - \mathbf {\overline t})^{\top} \right)\mathbf v_1
\\ &=\mathbf v_1^{\top}\mathbf S \mathbf v_1
\end{aligned}$$ (PCA1)


上式中，$\mathbf S$ 是协方差矩阵，由于存在约束条件 $\mathbf v_1^{\top} \mathbf v_1=1$，使用拉格朗日乘子法进行优化，

$$\mathcal L=V_1+ \lambda_1(1-\mathbf v_1^{\top} \mathbf v_1)=\mathbf v_1^{\top}\mathbf S \mathbf v_1+\lambda_1(1-\mathbf v_1^{\top} \mathbf v_1)$$

求梯度并令其为零，

$$\nabla \mathcal L=2 \mathbf {S v}_1-2\lambda_1 \mathbf v_1=\mathbf 0$$

得到

$$\mathbf {S v}_1=\lambda_1 \mathbf v_1$$


$$L=\mathbf v_1^{\top}\mathbf S \mathbf v_1=\lambda_1 \mathbf v_1^{\top} \mathbf v_1=\lambda_1$$

所以方差最大为 $\mathbf S$ 的最大特征值，其对应的特征向量就是这里的第一主成分方向 $\mathbf v_1$。

**第 k 主成分**

寻找第 $k$ 主成分方向 $\mathbf v_k, \ k > 1$，此时多了 $k$ 个约束条件 $\mathbf v_i^{\top} \mathbf v_k=0, \ i=1,2,\cdots , k-1$，拉格朗日函数为

$$\mathbf L = \mathbf v_k^{\top}\mathbf S \mathbf v_k +\lambda_k(1-\mathbf v_k^{\top} \mathbf v_k)+\sum_{i=1}^{k-1}\lambda_i \mathbf v_i^{\top} \mathbf v_k$$

取梯度并令其为零，

$$\nabla \mathcal L = 2 \mathbf {S v}_k-2\lambda_k \mathbf v_k+\sum_{i=1}^{k-1}\lambda_i \mathbf v_i= \mathbf 0$$ (PCA2)



上式两边乘以 $\mathbf v_i^{\top}, \ (\forall i < k)$ 得

$$\begin{aligned} & \ \mathbf v_i^{\top} \mathbf {S v}_k - \lambda_k \mathbf v_i^{\top} \mathbf v_k + \sum_{j=1}^{k-1}\lambda_j  \mathbf v_i \mathbf v_j=0
\\ \Rightarrow & \ \mathbf v_i^{\top} \mathbf {S v}_k - \lambda_i=0
\\ \Rightarrow & \ \lambda_i= \mathbf v_i^{\top} \mathbf {S v}_k=\mathbf v_i^{\top} \mathbf S^{\top} \mathbf v_k=t_i \mathbf v_i^{\top} \mathbf v_k = 0
\end{aligned}$$

上面第二个 $\Rightarrow$ 符号右侧中的推导是因为 $\mathbf S$ 是对称阵，另外需要注意 $\lambda$ 是拉格朗日乘子的因子，为了避免混淆，特征值使用了 $t_i$。于是可知，$\lambda_i=0, \ \forall i < k$，代入 {eq}`PCA2` 得 $\mathbf {S v}_k=\lambda_k \mathbf v_k$，所以 $\mathbf v_k$ 也是 $\mathbf S$ 的特征向量，其特征值为 $\lambda_k$，此即数据在 $\mathbf v_k$ 方向上的投影方差。
